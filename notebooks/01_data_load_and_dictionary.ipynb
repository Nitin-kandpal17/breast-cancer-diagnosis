## üìå Task: Load UCI Breast Cancer Dataset + Create Data Dictionary

**Objective:** Load the dataset, explore basic statistics, and generate an Excel-based data dictionary.

**Dataset:** [Breast Cancer Wisconsin (Diagnostic)](https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+(Diagnostic))  
**Source:** sklearn.datasets (preprocessed version)

---


```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

# Load dataset
data = load_breast_cancer()

# Create dataframe with features
df = pd.DataFrame(data.data, columns=data.feature_names)

# Add target
df["target"] = data.target

# Preview
df.head()
```

### üîç Dataset Information

- **Rows (samples):** 569
- **Features:** 30
- **Target values:** 0 (Malignant), 1 (Benign)

The dataset describes characteristics of the cell nuclei present in digitized image of a breast mass.

---

**Feature Groups:**
- **Mean values**
- **Standard Error**
- **Worst (largest) values**

For example:
- `mean radius`, `mean texture`, `mean area` ...
- `worst perimeter`, `worst concave points`, etc.

---


```python
# Create data dictionary
desc = df.describe().T
desc["null_count"] = df.isnull().sum()
desc["data_type"] = df.dtypes

# Save as Excel
desc.to_excel("../outputs/data_dictionary.xlsx")

desc.head()
```
